```python
import torch
import captum
from dataset import make_dataloaders
from easydict import EasyDict  as edict

task_args = edict()
play_args = edict()
eval_args = edict()
args = edict()
args.dataset_paths = {'synmnist': "SynMNIST"}
task_args.mode = ['std']
args.aid_perc = 0.0              # What ratio should we add. 0: same as baseline, 1: every image has a copy
task_args.dataset = {'name': 'synmnist', 'p': 0.95 , 'bg': 'nobg', 'splits': ['train','val','test'], 'baseline': False, 'bs': 9}
play_args.dataset = {'synmnist': { 'p': 0.95 , 'bg': 'nobg', 'splits': ['test'], 'bs': 128}}


for k,v in task_args.items():
    args['task_'+k] = v
for k,v in play_args.items():
    args['play_'+k] = v
for k,v in eval_args.items():
    args['eval_'+k] = v
    
args.eval_datasets = dict()
args.task_args = task_args
#dls = make_dataloaders(args)
```

```python
import torch.nn as nn
from os.path import join
from captum.attr import IntegratedGradients, Saliency
from models import simplecnn
import torch
import torchvision
import matplotlib.pyplot as plt
from torchvision.transforms import Normalize
import numpy as np
# Load the MNIST dataset

#test_dataset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=torchvision.transforms.ToTensor())

#envs = ['train', 'play']
probs = [0.5, 0.625]
probs = [0.5,0.625, 0.75,0.875, 0.95, 1]
envs = ['nobg','gradient','images']

envs = ['nobg']
splits = ['train',
          #'val',
          'test']
n_splits = len(splits)
n_probs = len(probs)

caption=""


# import a model
model_root = "models"
model_name = f"scnn_synmnist_0.75_nobg_bs_cmnist_baseline.pth"
model = simplecnn(args)
model_weights = torch.load(join(model_root,model_name))
model.load_state_dict(model_weights)
ig = IntegratedGradients(model)
saliency = Saliency(model)

t = Normalize([-1] * 3, [2] * 3)
t2 = Normalize([0.5], [1])
model.eval()
env = 'images'
colors=["Red","Green","Blue"]
    
fig, axs = plt.subplots(n_probs, 5, figsize=(n_probs*2, 4*5), constrained_layout=True)
args.task_dataset['bg'] = env 
for i, p in enumerate(probs):    

    args.task_dataset['p'] = p
    print(f'LOADING P ={p} , ENV={env}')
    dls = make_dataloaders(args)
    for j, split in enumerate(splits):

        x, y = next(iter(dls['task'][split]))
        # Make tensor require grad
        for param in x:
            param.requires_grad = True
        baseline = torch.zeros(x.shape)

        attributions, delta = ig.attribute(x, baseline, target=0, return_convergence_delta=True)
        grads = saliency.attribute(x)
        #grads = np.transpose(grads.squeeze().cpu().detach().numpy(), (1, 2, 0))

        for j in range(3):
            grid = torchvision.utils.make_grid(3*attributions[:,j].unsqueeze(1) , nrow=int(x.shape[0]**0.5))
            axs[i,j].imshow(grid.permute(1,2,0))
            axs[i,j].axis('off')
            if i == 0:
                axs[i,j].set_title(colors[j])
                axs[i,-1].set_title("Original Image")
            if j == 0:
                axs[i,j].axis('on')
                axs[i,j].xaxis.set_visible(False)

                axs[i,j].set_ylabel(p)

        grid2 = torchvision.utils.make_grid(t(x)[:,0].unsqueeze(1), nrow=int(x.shape[0]**0.5))

        grid3 = torchvision.utils.make_grid(t(x), nrow=int(x.shape[0]**0.5))
        axs[i,-2].imshow(grid2.permute(1,2,0))
        axs[i,-2].set_title("Red Channel")
        axs[i,-2].axis('off')
        axs[i,-1].imshow(grid3.permute(1,2,0))
        axs[i,-1].axis('off')


fig.savefig(f'SynMNIST_{env}_attributions.png')
```

```python
def get_grads(model):
    grads = dict()
    for name, param in model.named_parameters():
        if param.grad is not None:
            grads[name] = param.grad.cpu()
    return grads

def add_grads(grad, new_grad):
    for n, g in grad.items():
        grad[n] += g
    
    return grad
```
